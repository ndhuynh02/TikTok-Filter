```python
from PIL import Image
from deepface.detectors import FaceDetector

img_path = 'data/DLIB/ibug_300W_large_face_landmark_dataset/afw/134212_1.jpg'
img = cv2.imread(img_path)

# choose one of these:
#         "opencv"
#         "ssd"
#         "dlib"
#         "mtcnn"
#         "retinaface"
#         "mediapipe"
detector_name = "mtcnn"

detector = FaceDetector.build_model(detector_name)

obj = FaceDetector.detect_faces(detector, detector_name, img)

print("there are ",len(obj)," faces")
```

```python
bbox = obj[1][1]
bbox
```

```python
image = cv2.rectangle(img, (bbox[0], bbox[1]), (bbox[0] + bbox[2], bbox[1] + bbox[3]), (0, 255, 0))
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```
